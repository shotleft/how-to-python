```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("sales-funnel.xlsx")
df.head()
```

```python
df["Status"] = df["Status"].astype("category")
```

```python
pd.pivot_table(df, index = ["Name"]) # At its most basic: a simple 1-index pivot
```

```python
pd.pivot_table(df,index=["Name","Rep","Manager"]) # You can also have multiple indices
```

```python
pd.pivot_table(df,index=["Manager","Rep"]) # A 2-index pivot
```

```python
pd.pivot_table(df,index=["Manager","Rep"], values = ["Price"]) # We don't want Account and Quantity, so let's 
                                                               # rather be explicit about which values (columns)
                                                               # we want to see
```

```python
# Note that in the above the values are being AVERAGED by default which is no good to man or beast!
```

```python
pd.pivot_table(df,index=["Manager","Rep"], values = ["Price"], aggfunc = np.sum) # i.e. the numpy sum function
```

```python
# aggfunc can take a variety of functions - in this example we obtain the average via np.mean (from numpy) 
# and the number of items via len (from python)
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc=[np.mean,len])
```

```python
# Now let's get sophisticated and add *columns*. A note from author @chris1610:
# I think one of the confusing points with the pivot_table is the use of columns and values . 
# Remember, columns are optional - they provide an additional way to segment the actual values you care about. 
# The aggregation functions are applied to the values you list.
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],
               columns=["Product"],aggfunc=[np.sum])
```

```python
# In the above, we don't like the NaN's very much, let's replace them with 0's using the fill_value parameter:
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],
               columns=["Product"],aggfunc=[np.sum], fill_value = 0)
```

```python
# Now let's complicate life and add Quantity into the mix in addition to Price
```

```python
pd.pivot_table(df, index = ["Manager", "Rep"], values = ["Price", "Quantity"],
              columns = ["Product"], aggfunc = [np.sum], fill_value = 0)
```

```python
# Items can be moved between the Index and the Columns to get a different view on things:
```

```python
pd.pivot_table(df, index = ["Manager", "Rep", "Product"], values = ["Price", "Quantity"], 
               aggfunc = np.sum, fill_value = 0)
```

```python
# Now we want to get some totals (and we've added both sum and mean into the mix) - use margins = True
```

```python
pd.pivot_table(df,index=["Manager","Rep","Product"],
               values=["Price","Quantity"],
               aggfunc=[np.sum,np.mean],fill_value=0,margins=True)
```

```python
# Now let's look at Manager level, a new pivot
```

```python
pd.pivot_table(df,index=["Manager","Status"],values=["Price"],
               aggfunc=[np.sum],fill_value=0,margins=True)
```

```python
# What is very useful is being able to pass aggfunc a dictionary, via which you can say what kind of function
# you want performed on each of the numeric values
```

```python
pd.pivot_table(df,index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"],
               aggfunc={"Quantity":len,"Price":np.sum},fill_value=0)
```

```python
# AND you can also supply more than one function to a value via a list nested within your dictionary:
```

```python
pd.pivot_table(df,index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"],
               aggfunc={"Quantity":len,"Price":[np.sum,np.mean]},fill_value=0)
```

```python
# So let's finalize this now by assigning our pivot to a variable
```

```python
table = pd.pivot_table(df,index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"],
               aggfunc={"Quantity":len,"Price":[np.sum,np.mean]},fill_value=0)
```
